````python
# Question 1
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.arange(0 , 11)
y = x
plt.figure(figsize = (6, 4))
plt.plot(x, y, 'black')
plt.xlabel('x')
plt.ylabel('y')
plt.title(r"Plot of $y=x$")
plt.show ()
````

````python
# Question2
x = np.arange(-10,11)
y = x**2
plt.figure(figsize=(5,6))
plt.plot(x,y,'r')
plt.xlabel('x')
plt.ylabel('y')
plt.title(r"plot of $y = x^2")
plt.show()
print(x)
````

````python
# Question 3
x = np.random.randint(0, 10, 50)
y = np.random.randint(0, 10, 50)
plt.scatter(x, y)
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Scatter Plot of 50 Random Points')
plt.show()

````

````python
# Question 4

x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
plt.plot(x, y)
plt.xlabel('x values (radians)')
plt.ylabel('sin(x)')
plt.title('Sine Wave')
plt.show()

````

````python
# Question 5

categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 9]

plt.bar(categories, values)


plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart of Categories and Values')
plt.show()
````

````python
# Question 6
data = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(data, bins=30, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 1000 Random Numbers from a Normal Distribution')
plt.show()


````

````python
# Question 7

x = np.linspace(0, 10, 100)
y1 = x**2    
y2 = x**3    

plt.plot(x, y1, label='y = x^2', color='blue')
plt.plot(x, y2, label='y = x^3', color='orange')
plt.xlabel('x values')
plt.ylabel('y values') 
plt.show()

````

````python
# Question 8
x = np.linspace(0, 10, 100)
y1 = x**2    
y2 = x**3    

plt.plot(x, y1, label='y = x^2', color='blue')
plt.plot(x, y2, label='y = x^3', color='orange')
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Line Plots of y = x^2 and y = x^3')
plt.legend()
plt.grid(True) 
plt.show()

````

````python
# Question 9
x = np.linspace(0, 10, 100)
y1 = x**2   
y2 = x**3    
fig, axs = plt.subplots(2, 1, figsize=(8, 10))  
axs[0].plot(x, y1, label='y = x^2', color='blue')
axs[0].set_title('Plot of y = x^2')
axs[0].set_xlabel('x values')
axs[0].set_ylabel('y values')
axs[0].legend()
axs[0].grid(True)  
axs[1].plot(x, y2, label='y = x^3', color='orange')
axs[1].set_title('Plot of y = x^3')
axs[1].set_xlabel('x values')
axs[1].set_ylabel('y values')
axs[1].legend()
axs[1].grid(True) 

plt.tight_layout()

plt.show()

````
